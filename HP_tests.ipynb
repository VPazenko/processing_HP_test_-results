# Load the data

The data was collected from a research among patients who came to the hospital (Russia, St. Petersburg) with the suspicion of HP infection. According to the study design, each patient was tested by 2 breath tests C13 UBT (Urea Breath Test) , as well as FGDS with biopsy sampling, which were analyzed histologically, by PCR and by RUT (Rapid Urea Test).

The data looks like a three-level table. The top level contains information about the biopsy sampling place:
- Pyloric antrum
- Incisura angularis
- Body (stomach)
- 13С BT Heliforce - Breath test (BT) with load Heliforce
- 13С BT Хеликарб - Breath test (BT) with load Helicarb 

Next level contains information about type of test:
- Result RUT = Rapid Urea Test
- PCR = Polymerase Chain Reaction
- Histology
- Breath test

Last one contains specific information about some tests:
For Histology:
- bacil = percentage of bacillary forms
- coccal = percentage of coccal forms
- status HP = result 
For BT:
- DOB 0 = delta over base at basal exhalation (before load)
- DOB 30 = delta over base at load exhalation (30 min after taking load)
- Δ = 'DOB 30' - 'DOB 0'
- Status = result

We also have an additional df that contains a recheck of some histologic samples by another scientist (structure is the same)

```python
config_dict = open_config_yaml('config.yaml')
research_df = pd.read_excel(config_dict['data'], header=[0,1,2])
hystology2_df = pd.read_excel(config_dict['recheck_hystology'], header=[0,1,2])
```

```python
#Rename column in second df
hystology2_df = hystology2_df.rename(columns={'Histology': 'Histology rechecked by another scientist'}, level=1)
#Merge them
research_df = research_df.merge(hystology2_df, on=None, how='left')
#Set index
index_key = ('Patient number', 'Unnamed: 0_level_1', 'Unnamed: 0_level_2',)
research_df = research_df.set_index(keys=index_key)
research_df.index.names = ['Patient number']
```

# Let's look at BT in more detail

```python
bt_index_location = [17, 18, 19, 20, 21, 22, 23, 24]
breath_test_df = research_df_merged.iloc[:,bt_index_location]
#we drop all rows, where less then 2 values (it means that in this row we have not enough values for easy recovery) 
breath_test_df = breath_test_df.dropna(thresh=2)
#Check for missing data
percent = 100 * (breath_test_df.isnull().sum() / len(breath_test_df))
print(round(percent, 2))
```

As we can see, we can easily recover some of the data in the last column based on Δ value (all Δ > 4 cause status '+')

```python
miss = '13С BT (Helicarb), status'
result = '13С BT (Helicarb), Δ'

df_missing = breath_test_df[breath_test_df[miss].isna()]
df_missing[miss] = df_missing[result].apply(lambda x: compare(x, 4, 4))
breath_test_df[miss] = breath_test_df[miss].fillna(df_missing[miss])
```

Let's select only columns with status, convert string values to numeric values and build a heatmap

```python
breath_test_status_df = breath_test_df.iloc[:,[3,7]]
breath_test_status_df = breath_test_status_df.replace(to_replace=[' +', '+', '-', ' -', '  -'], value=[1, 1, 0, 0, 0])

create_heatmap(breath_test_status_df)
```

As we can see, the correlation between these two columns is quite strong. 

However, looking at the detailed instructions for the tests, we can see that the result is based on 'Δ'. It can also be noticed that there is some grey zone where the result cannot be interpreted definitely (it is recommended to repeat the test or use another way of detecting HP).

Let's change the status of both tests based on 'Δ' and manufacturer's recommendations. (Grey zone will score 0.5)

```python
breath_test_df['13С BT (Helicarb), status'] = breath_test_df.iloc[:,6].map(lambda x: compare(x, 4.5, 3.5))
breath_test_df['13С BT (Heliforce), status'] = breath_test_df.iloc[:,2].map(lambda x: compare(x, 4.4, 3.6))
breath_test_status_df2 = breath_test_df.iloc[:,[3,7]]

create_heatmap(breath_test_status_df2)
```

The correlation has gotten stronger.

```python
breath_test_status_df3 = breath_test_df.iloc[:,[3,7]]
breath_test_status_df3 = breath_test_status_df3[breath_test_status_df3 != 0.5]

create_heatmap(breath_test_status_df3)
```

# Let's compare the 'gold standard' and the BT

```python
elem_list = list(hist_df_status.columns)
#Let's add one more column 'status'. The status is considered positive if at least one region of the stomach is detected with HP
hist_df_status['status'] = hist_df_status.apply(lambda x: compare_results(x, elem_list, how='status'), axis=1)

#Merge Histology and BT
bt_versus_hist = pd.merge(hist_df_status, breath_test_status_df3, left_index=True, right_index=True)

create_heatmap(bt_versus_hist)
```

As we can see, the correlation of the tests is insignificant (20%).

**However, there is an explanation for this point!** 

The breath test is based on the enzymatic reaction of urea (with marked C13) hydrolysis under the action of the enzyme urease (this enzyme produces HP). If we look at the histology data, we can see that in some cases only coccal forms have been identified, which cannot produce urease. 

Let's redo the comparison based on the fact that bacillary forms must be greater than 0 to detect HP.

```python
name = 'bacil'
elem_list = []
columns = list(research_df_merged.columns)
for elem in columns:
    result = re.search(r'bacil', elem) and not re.search(r'recheck', elem)
    if result:
        elem_list.append(elem)
elem_list

hist_df_old = research_df_merged.loc[:,elem_list].copy()

#create a labels based on amount if bacil forms
hist_df_old2 = create_result(hist_df_old, list_com, value='amount_0')
hist_df_old2 = hist_df_old2.iloc[:,5:]

#Change labels for numbers
hist_df_old2 = hist_df_old2.replace(to_replace=['-', '+'], value=[0, 1])
elem_list = list(hist_df_old2.columns)

#Common column creation
hist_df_old2['status'] = hist_df_old2.apply(lambda x: compare_results(x, elem_list, how='status'), axis=1)

#Add BT with all data
breath_test_df = research_df_merged.iloc[:,bt_index_location]
breath_test_df['13С BT (Helicarb), status'] = breath_test_df.iloc[:,6].map(lambda x: compare(x, 4.5, 3.5))
breath_test_df['13С BT (Heliforce), status'] = breath_test_df.iloc[:,2].map(lambda x: compare(x, 4.4, 3.6))

bt_versus_hist2 = pd.merge(hist_df_old2, breath_test_df.iloc[:,[3,7]], left_index=True, right_index=True)
create_heatmap(bt_versus_hist2)
```

It can be noticed that the correlation has improved significantly.
